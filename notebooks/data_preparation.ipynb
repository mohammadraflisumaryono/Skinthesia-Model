## Agregasi Data

```python
import pandas as pd
from collections import Counter

# Fungsi untuk gabungkan list dan hilangkan duplikat
def merge_unique(series):
    combined = set()
    for item in series.dropna():
        if isinstance(item, str):
            try:
                parsed = eval(item) if item.startswith('[') else [item]
            except:
                parsed = [item]
            combined.update(parsed)
        elif isinstance(item, list):
            combined.update(item)
    return list(combined)

# Fungsi untuk ambil modus (nilai terbanyak)
def get_mode(series):
    try:
        return series.mode().iloc[0]
    except:
        return None

# Agregasi per URL
agg_df = df.groupby('url').agg({
    'product_name': 'first',
    'brand': 'first',
    'category': 'first',
    'price': 'first',
    'rating': 'first',
    'skin_type': merge_unique,
    'total_reviews': 'sum',
    'std_skin_type': merge_unique,
    'std_skin_concern': merge_unique,
    'std_ingredients': merge_unique,
    'std_skin_goal': merge_unique,
    'age': get_mode,
    'rating_star': 'mean'
}).reset_index()
```

## Data Cleaning

Ubah Tipe Data

```python
import re

def clean_price(price):
    if isinstance(price, str):
        # Hapus semua non-digit (angka) dan non-koma atau titik
        # Tapi sebelumnya hilangkan "Rp" dan spasi
        price = price.replace('Rp', '').replace('rp', '').strip()

        # Hapus titik sebagai pemisah ribuan
        price = price.replace('.', '')

        # Ganti koma jadi titik jika memang ada (desimal)
        price = price.replace(',', '.')

        # Ambil hanya digit dan titik desimal
        cleaned = re.sub(r'[^\d.]', '', price)

        try:
            return float(cleaned)
        except:
            return None
    return price

cleaned_df['price'] = cleaned_df['price'].apply(clean_price)
```

```python
cleaned_df.info()
```

Gabungkan skin_type dan found_skin_type

```python
import ast
import pandas as pd

def to_list(value):
    if isinstance(value, float) and pd.isna(value):
        return []
    if isinstance(value, list):
        return value
    if isinstance(value, str):
        try:
            parsed = ast.literal_eval(value)
            if isinstance(parsed, list):
                return parsed
            else:
                return [parsed]
        except:
            return [value]
    return [value]

def combine_skin_types(row):
    user_skin = to_list(row.get('skin_type'))
    found_skin = to_list(row.get('std_skin_type'))
    combined = list(set(user_skin + found_skin))
    return combined

# Jalankan
cleaned_df['combined_skin_type'] = cleaned_df.apply(combine_skin_types, axis=1)
```

```python
cleaned_df.drop(columns=['skin_type','std_skin_type'],inplace=True)
```

```python
cleaned_df.info()
```

Hapus value duplikat

```python
def remove_duplicates(x):
    if isinstance(x, list):
        # Ubah semua elemen jadi lowercase
        lowercased = [str(item).lower() for item in x]
        # Hapus duplikat dengan set lalu kembalikan ke list
        return list(set(lowercased))
    return x

# Buat kolom baru dengan versi bersih
cleaned_df['clean_skin_concern'] = cleaned_df['std_skin_concern'].apply(remove_duplicates)
cleaned_df['clean_ingredients'] = cleaned_df['std_ingredients'].apply(remove_duplicates)
cleaned_df['clean_skin_goal'] = cleaned_df['std_skin_goal'].apply(remove_duplicates)
cleaned_df['clean_skin_type'] = cleaned_df['combined_skin_type'].apply(remove_duplicates)

# Hapus kolom lama
cleaned_df.drop(columns=['std_skin_concern', 'std_ingredients', 'std_skin_goal', 'combined_skin_type'], inplace=True)
```

## Encode Data

```python
encoded_df = cleaned_df.copy()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer

# Buat binarizer untuk setiap fitur multi-label
mlb_ingredients = MultiLabelBinarizer()
mlb_concerns = MultiLabelBinarizer()
mlb_types = MultiLabelBinarizer()
mlb_goals = MultiLabelBinarizer()

# Fit dan transform lalu konversi ke DataFrame
df_ingredients = pd.DataFrame(
    mlb_ingredients.fit_transform(encoded_df['clean_ingredients']),
    columns=[f"ingredient_{c}" for c in mlb_ingredients.classes_],
    index=encoded_df.index
)

df_concerns = pd.DataFrame(
    mlb_concerns.fit_transform(encoded_df['clean_skin_concern']),
    columns=[f"concern_{c}" for c in mlb_concerns.classes_],
    index=encoded_df.index
)

df_types = pd.DataFrame(
    mlb_types.fit_transform(encoded_df['clean_skin_type']),
    columns=[f"type_{c}" for c in mlb_types.classes_],
    index=encoded_df.index
)

df_goals = pd.DataFrame(
    mlb_goals.fit_transform(encoded_df['clean_skin_goal']),
    columns=[f"goal_{c}" for c in mlb_goals.classes_],
    index=encoded_df.index
)

# Gabungkan ke DataFrame asli dan hapus kolom aslinya
encoded_df = pd.concat([
    encoded_df.drop(columns=[
        'clean_ingredients', 'clean_skin_concern', 'clean_skin_type', 'clean_skin_goal'
    ]),
    df_ingredients, df_concerns, df_types, df_goals
], axis=1)
```
